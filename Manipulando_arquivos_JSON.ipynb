```python
#Importando pandas
import pandas as pd
```

```python
#Lendo arquivos json
dados_pacientes = pd.read_json('/content/pacientes.json')
```

```python
#Exibindo os Dados
dados_pacientes
```

```python
#Lendo arquivos json e armazenando
dados_pacientes2 = pd.read_json('/content/pacientes_2.json')
```

```python
#Exibido Dados
dados_pacientes2
```

```python
#Armazenando Dados JSON normalizados
df_normalizado = pd.json_normalize(dados_pacientes2['Pacientes'])
#               |--------a--------||-------b-------||-----c-----|
# a = metodo para normalizar JSON aninhados
# b = variável onde estão os Dados
# c = Coluna onde estão os Dados aninhados
```

```python
#Exibindo DataFrame
df_normalizado
```

```python
# Copiando Dados JSON normalizados
df_normalizado.to_json('Historico_pacientes_normalizado.json')
```

```python
import requests # biblioteca para requisição de API
import json # biblioteca para ler Dados
import pandas as pd # biblioteca para manipular Dataframe
```

```python
# Armazenando Dados JSON com API
dados = requests.get('https://jsonplaceholder.typicode.com/users')
```

```python
# Recuperando Dados JSON da URL
visualizacao = json.loads(dados.text)
visualizacao # Exibindo JSON
```

```python
# Armazenando DataFrame
df_cadastro = pd.DataFrame(visualizacao)
df_cadastro # Exibindo DataFrame
```

```python
# Utilizando metodo para normalizar com separador undescore
normalizando = pd.json_normalize(visualizacao,sep='_')
normalizando # Exibindo DataFrame normalizado
```
