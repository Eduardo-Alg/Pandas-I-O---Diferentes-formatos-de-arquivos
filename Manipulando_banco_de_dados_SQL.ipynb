```python
#importando biblioteca
import sqlalchemy
```

```python
#importando objetos da biblioteca
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
#Criando banco de Dados local
engine = create_engine('sqlite:///:memory')
```

```python
# Pegando URL dos Dados
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
# Importando biblioteca
import pandas as pd
```

```python
# Lendo arquivo csv
dados = pd.read_csv(url)
```

```python
# Exibindo DataFrame
dados.head()
```

```python
# Escrevendo Banco de Dados chamado clientes
dados.to_sql('novos_clientes',engine, index=False)
```

```python
# Visualização de metadados
inspector = inspect(engine)
```

```python
# Imprime o Banco de Dados
print(inspector.get_table_names())
```

```python
# Consulta SQL ao banco de Dados
query = 'SELECT * FROM novos_clientes WHERE Categoria_de_renda="Empregado"'
#       |---a---| |-----b-----||---------------c-------------------|
# a = Pega todas as colunas com *
# b = De qual tabela
# c = Aplica filtro
```

```python
# Lendo o resultado da consulta
empregados = pd.read_sql(query, engine)
```

```python
# Exibindo DataFrame da consulta
empregados
```

```python
# Salvando resultado como nova tabela
empregados.to_sql('clientes_empregados',con=engine, index=False)
```

```python
# Ler tabela inteira
pd.read_sql_table('clientes_empregados', engine)
```

```python
# Lendo algumas colunas especificadas
pd.read_sql_table('clientes_empregados', engine, columns=['ID_Cliente','Grau_escolaridade','Rendimento_anual'])
```

```python
# Imprime tabelas criadas
print(inspector.get_table_names())
```

```python
pd.read_sql_table('clientes_empregados',engine)
```

```python
print(inspector.get_table_names())
```

```python
# Função para deletar registro no banco de dados
query = 'DELETE FROM novos_clientes WHERE ID_Cliente=5008804	'
#|-a-|  |---------------------b----------------------------|
# a = variável com a clausula para DELETAR registro
# b = Cláusula SQL para deletar ID_Cliente=5008804
with engine.connect() as conn: # Gerenciamento automatico de abertua e fechamento de conexão.
  result = conn.execute(text(query)) # executa a cláusula a função text garante que seja interpretada como SQL puro
  conn.commit() # Confirma a transação
```
