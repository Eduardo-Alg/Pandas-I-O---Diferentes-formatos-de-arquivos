```python
#Importando biblioteca Pandas
import pandas as pd
```

```python
#Armazenando os Dados brutos na variável (?raw=True) para carregar os Dados brutos
url = 'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True'
```

```python
#Lendo os Dados armazenados na url e armazenando em ua variável
dados_co2 = pd.read_excel(url)
```

```python
#Exibindo as primeiras linhas
dados_co2.head()
```

```python
#Metodo para verificar se a planila .xlsx(Excel) tem outras páginas
pd.ExcelFile(url).sheet_names
```

```python
#Armazenando os Dados na página 'emissoes_percapita'
percapita = pd.read_excel(url,sheet_name='emissoes_percapita')
```

```python
#Exibindo os Dados na página referenciada no parâmetro
percapita.head()
```

```python
#Armazenando os Dados na página 'fontes'
fontes = pd.read_excel(url, sheet_name='fontes')
```

```python
#Exibe as 5 primeiras linhas dos Dataframe
fontes.head()
```

```python
#Armazena os Dados da página 'emissoes_C02' especificando as colunas que serão salvas 'A:D'
intervalo = pd.read_excel(url , sheet_name='emissoes_C02', usecols='A:D')
```

```python
#Exibe a variável
intervalo
```

```python
#Passando mais um parâmetro armazenando as colunas que serão salvas
intervalo_2 = pd.read_excel(url , sheet_name='emissoes_C02', usecols='A:D', nrows=10)
```

```python
#Exibe a variável com os parâmetro de colunas e linhas selecionadas
intervalo_2
```
